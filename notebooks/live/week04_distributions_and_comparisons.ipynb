## Common mistakes

- **Bar chart of means for distributions.** It hides spread and outliers entirely.
- **Too few or too many histogram bins.** Both distort shape; try a few bin counts.
- **Ignoring skew.** Reporting a mean income on a right-skewed variable overstates the typical customer; prefer the median or a log scale.
