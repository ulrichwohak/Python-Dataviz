## Why visualise?

A table of 7,000 weekly sales rows hides the story; a single line shows the trend at a glance. Every chart **encodes** data values as visual properties — position, length, colour. Good charts pick encodings the eye reads accurately: *position* and *length* are read best, which is why lines and bars dominate business reporting.

## Common mistakes

- **Unsorted bars.** Leaving regions in data order makes the ranking hard to read — always sort categorical bars by value.
- **Truncated y-axis.** Starting a bar chart's axis above zero exaggerates differences.
- **Line chart for unordered categories.** A line implies continuity; never use one to compare regions.
