## Common mistakes

- **Mapping totals, not rates.** A choropleth of total sales mostly maps population; use a per-capita rate.
- **Measuring area in degrees.** Always reproject to a projected CRS first.
- **Too many colour classes.** More than ~6 bins are hard to distinguish; pick a classification scheme deliberately.
