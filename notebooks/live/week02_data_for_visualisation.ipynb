## Tidy data

Most plotting bugs are really *data shape* bugs. **Tidy data** means one row per observation and one column per variable. `groupby` collapses rows to a summary; `pivot_table` reshapes long data into a grid (e.g. month × category) that maps directly onto a multi-series chart. Missing values must be handled *before* plotting, or they silently distort it.

## Common mistakes

- **Plotting before aggregating.** Passing 8,000 raw rows to a line plot yields a spiky mess; summarise first.
- **`mean` vs `sum`.** `pivot_table` defaults to `mean`; pick the aggregation on purpose.
- **Silent NaNs.** Some plots drop missing values without telling you — always count them yourself first.
