## From static to interactive

Static charts make one point; **interactive** charts let a stakeholder ask their own follow-up questions — hover for exact values, zoom into a period, toggle series. `plotly` builds these with little code. When several charts and filters belong together, a **dashboard** (we use Streamlit) is the natural home — see `dashboards/app.py`.

## Load the data

```python
dash = load("dashboard")
dash.head()
```

### An interactive line with plotly

Total monthly revenue, with hover tooltips and zoom for free.

```python
import plotly.express as px

monthly = dash.groupby("month", as_index=False)["revenue"].sum()
fig = px.line(monthly, x="month", y="revenue",
              title="Monthly revenue (hover and zoom)")
fig
```

### An interactive bar with a colour dimension

Revenue by region; click legend entries to filter, hover for values.

```python
region_rev = dash.groupby("region", as_index=False)["revenue"].sum()
fig = px.bar(region_rev, x="region", y="revenue", color="region",
             title="Revenue by region")
fig
```

### A static export for the report

Dashboards live in the browser, but reports need a static image — so we also save a matplotlib version to `figures/`.

```python
monthly_region = dash.pivot_table(index="month", columns="region",
                                  values="revenue", aggfunc="sum")

fig, ax = plt.subplots()
monthly_region.plot(ax=ax)
ax.set_title("Monthly revenue by region")
ax.set_ylabel("Revenue (€)")
save_figure(fig, "week07_monthly_by_region")
plt.show()
```

## Common mistakes

- **Interactivity for its own sake.** If the message is one number, a static chart is clearer and more portable.
- **No default view.** An interactive chart should already tell the story before the user touches it.
- **Shipping a notebook as a dashboard.** Use a real app (Streamlit) so non-coders can use it — see the next note.

## Verification checklist for AI output

Before relying on any chart or code an assistant gives you, confirm:

1. Does the chart actually answer *how revenue moves over time and across regions*?
2. Are the axes, units, and scale correct and honest?
3. Is this the right chart type for the data?
4. Did the code run on **our** dataset (correct columns, no invented data)?
5. Were rows silently dropped, duplicated, or mis-aggregated?
6. Could the colour, ordering, or missing context mislead a reader?

If you cannot answer all six, fix it before moving on.
