## Beyond the average

A single mean hides the story. **Histograms** show one distribution's shape; **box plots** compare many groups compactly (median, quartiles, outliers); **violin plots** add the full density, revealing bimodality a box hides. Skewed variables like income often need a **log scale** to be readable.

## Load the data

```python
ops = load("operations")
customers = load("customers")
ops.head()
```

### A histogram for one distribution

Look at the shape of monthly performance — where is it centred and is it symmetric?

```python
fig, ax = plt.subplots()
ax.hist(ops["monthly_performance"], bins=30, color="#4C72B0")
ax.set_title("Distribution of monthly performance")
ax.set_xlabel("Performance score")
ax.set_ylabel("Employees")
plt.show()
```

### Box and violin plots for group comparison

Compare departments. The box plot is compact; the violin shows the full density. Note the outliers the box plot flags.

```python
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
sns.boxplot(data=ops, x="department", y="monthly_performance", ax=axes[0])
axes[0].set_title("Performance by department (box)")
sns.violinplot(data=ops.dropna(subset=["satisfaction_score"]),
               x="department", y="satisfaction_score", ax=axes[1])
axes[1].set_title("Satisfaction by department (violin)")
for a in axes:
    a.tick_params(axis="x", rotation=20)
save_figure(fig, "week04_dept_distributions")
plt.show()
```

### Taming skew with a log scale

Customer income is strongly right-skewed. A log scale spreads the bulk out so the distribution becomes readable.

```python
inc = customers["income"].dropna()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(inc, bins=40, color="#DD8452")
axes[0].set_title("Income (raw, right-skewed)")
axes[1].hist(np.log10(inc), bins=40, color="#DD8452")
axes[1].set_title("Income (log10 scale)")
axes[1].set_xlabel("log10(income)")
plt.show()
```

## Common mistakes

- **Bar chart of means for distributions.** It hides spread and outliers entirely.
- **Too few or too many histogram bins.** Both distort shape; try a few bin counts.
- **Ignoring skew.** Reporting a mean income on a right-skewed variable overstates the typical customer; prefer the median or a log scale.
