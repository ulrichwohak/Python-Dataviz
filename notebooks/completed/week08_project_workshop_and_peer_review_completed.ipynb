## Load the data

```python
customers = load("customers")
segment_summary = (customers.groupby("segment", as_index=False)
                   .agg(customers=("customer_id", "count"),
                        avg_clv=("customer_lifetime_value", "mean"),
                        avg_churn=("churn_probability", "mean")))
segment_summary
```

### One decision-ready chart

Average lifetime value by segment, ordered Value → Premium, with the Premium bar highlighted and values labelled.

```python
order = ["Value", "Mainstream", "Premium"]
summary = segment_summary.set_index("segment").loc[order]
colors = ["#B0B0B0", "#B0B0B0", "#55A868"]

fig, ax = plt.subplots()
bars = ax.bar(summary.index, summary["avg_clv"], color=colors)
for bar, value in zip(bars, summary["avg_clv"], strict=True):
    ax.text(bar.get_x() + bar.get_width() / 2, value, f"€{value:,.0f}",
            ha="center", va="bottom", fontsize=10)
ax.set_title("Premium customers are worth the most over their lifetime")
ax.set_ylabel("Average lifetime value (€)")
ax.margins(y=0.15)
save_figure(fig, "week08_clv_by_segment")
plt.show()
```

## Pause and predict

Your title claims Premium customers are worth the most. Is that claim safe? Premium is also the smallest segment — does *average* CLV alone justify a strategy decision, or do you need the segment sizes and churn next to it?

## Common mistakes

- **Title over-claims.** Make sure the data supports the sentence in the title.
- **One metric, big decision.** Average CLV ignores segment size and churn; show what the decision actually needs.
- **Skipping your own verification.** Run the checklist on your *own* chart, not just on AI output.

## Peer-review rubric

Score a peer's chart 1–5 on each: (a) right chart for the question, (b) honest axes and scale, (c) title states a supported takeaway, (d) clutter removed, (e) a reader could act on it. Give one concrete improvement.

## Verification checklist for AI output

Before relying on any chart or code an assistant gives you, confirm:

1. Does the chart actually answer *which customer segment is most valuable*?
2. Are the axes, units, and scale correct and honest?
3. Is this the right chart type for the data?
4. Did the code run on **our** dataset (correct columns, no invented data)?
5. Were rows silently dropped, duplicated, or mis-aggregated?
6. Could the colour, ordering, or missing context mislead a reader?

If you cannot answer all six, fix it before moving on.
