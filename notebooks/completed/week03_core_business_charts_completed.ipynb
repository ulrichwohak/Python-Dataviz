## The core four

Most business reporting is line charts (trend), grouped bars (compare levels), stacked bars (compose a total), and small multiples (compare many series without overplotting). Each answers a different question; choosing wrongly is the most common charting error in industry.

## Load the data

```python
sales = load("sales")
sales = sales.dropna(subset=["revenue"])
sales.head()
```

### Multi-series line by channel

Aggregate to monthly revenue per channel and draw one line each.

```python
m = sales.assign(month=sales["date"].dt.to_period("M").dt.to_timestamp())
by_channel = m.pivot_table(index="month", columns="channel",
                           values="revenue", aggfunc="sum")

fig, ax = plt.subplots()
by_channel.plot(ax=ax)
ax.set_title("Monthly revenue by channel")
ax.set_ylabel("Revenue (€)")
plt.show()
```

### Grouped vs stacked bars

Grouped bars compare regions *within* a channel; stacked bars show how channels compose each region's total. Same data, different question.

```python
grp = sales.groupby(["region", "channel"])["revenue"].sum().unstack()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
grp.plot(kind="bar", ax=axes[0])
axes[0].set_title("Grouped: compare regions")
grp.plot(kind="bar", stacked=True, ax=axes[1])
axes[1].set_title("Stacked: compose the total")
for a in axes:
    a.set_ylabel("Revenue (€)")
    a.tick_params(axis="x", rotation=0)
save_figure(fig, "week03_grouped_vs_stacked")
plt.show()
```

### Small multiples

When series overlap too much, give each its own panel. Seaborn's `relplot` with `col=` builds the grid for us.

```python
import seaborn as sns

monthly = (sales.assign(month=sales["date"].dt.to_period("M").dt.to_timestamp())
           .groupby(["month", "product_category"], as_index=False)["revenue"].sum())
g = sns.relplot(data=monthly, x="month", y="revenue", col="product_category",
                col_wrap=2, kind="line", height=2.8)
g.set_axis_labels("Month", "Revenue (€)")
plt.show()
```

## Common mistakes

- **Stacking to compare.** Stacked bars hide individual series lengths above the first; use grouped bars to compare.
- **Spaghetti charts.** Too many overlapping lines — switch to small multiples.
- **Inconsistent axes across panels.** Small multiples must share a y-scale to be comparable (seaborn does this by default).

## Verification checklist for AI output

Before relying on any chart or code an assistant gives you, confirm:

1. Does the chart actually answer *how channels and regions contribute to revenue*?
2. Are the axes, units, and scale correct and honest?
3. Is this the right chart type for the data?
4. Did the code run on **our** dataset (correct columns, no invented data)?
5. Were rows silently dropped, duplicated, or mis-aggregated?
6. Could the colour, ordering, or missing context mislead a reader?

If you cannot answer all six, fix it before moving on.
